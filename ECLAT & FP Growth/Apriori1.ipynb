```python
df = pd.read_csv("C:/Users/SCIENCE/Documents/market1.csv")
df
```

```python
items=df.columns
items
```

```python
encoded_vals = []
for index, row in df.iterrows(): 
    labels = {}
    uncommons = list(set(items) - set(row))
    commons = list(set(items).intersection(row))
    for uc in uncommons:
        labels[uc] = 0
    for com in commons:
        labels[com] = 1
    encoded_vals.append(labels)
```

```python
ohe_df = pd.DataFrame(encoded_vals)
ohe_df 
```

```python
freq_items = apriori(ohe_df, min_support = 0.002, use_colnames = True, verbose = 1)
freq_items
```

```python
sort1 = association_rules(freq_items, metric = "confidence", min_threshold = 0.006)
sort1.sort_values("confidence", ascending= False)
```

```python
p=association_rules(freq_items, metric = "confidence", min_threshold = 0.006)
Newconfidence=p["confidence"] * 100
Newconfidence
```

```python
newlift=np.round(p["lift"])
newlift
```
