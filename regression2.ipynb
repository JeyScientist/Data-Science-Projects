```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
df = pd.read_csv("C:/Users/SCIENCE/Documents/mart1.csv")
df
```

```python
df.describe()
```

```python
df.dropna()
df.isnull().sum()
```

```python
df["Item_Weight"]=df["Item_Weight"].replace([np.nan,0])
df["content"] = df["content"].replace([np.nan,0])
df.isnull().sum()
```

```python
y = df["Item_Weight"]
x = df.drop("Item_Weight", axis = 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
df.info()
```

```python
rf = RandomForestRegressor(n_estimators = 301)
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)
rf_pred
```

```python
print(mean_squared_error(rf_pred, y_test))
```

```python
sv = SVR()
sv.fit(x_train, y_train)
sv.pred = sv.predict(x_test)
sv.pred
```

```python
print(mean_squared_error(sv.pred, y_test))
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
lr.pred = lr.predict(x_test)
lr.pred
```

```python
print(mean_squared_error(lr.pred, y_test))
```

```python
la = Lasso()
la.fit(x_train, y_train)
la.pred = la.predict(x_test)
la.pred
```

```python
print(mean_squared_error(la.pred, y_test))
```

```python
rd = Ridge(alpha =1)
rd.fit(x_train, y_train)
rd.pred = rd.predict(x_test)
rd.pred
```

```python
print(mean_squared_error(rd.pred, y_test))
```
